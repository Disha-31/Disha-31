```python
avg_tons = data.groupby('COUNTRY_FLYING_MISSION')['TONS_HE', 'TONS_IC','TONS_FRAG'].sum()
print(avg_tons)
```

```python
#As we have more than 170,000 entries we randomly sample 50rows
#sample = df.sample(50)
source = ColumnDataSource(avg_tons)
```

# Feature 1 : Stacked bar charts with interactive legend

```python
# Import the desired palette from bokeh.palettes library
# Here, we are importing Viridis palette.
from bokeh.palettes import Viridis
output_notebook()

country = source.data['COUNTRY_FLYING_MISSION'].tolist()

# Create the figure 
p1 = figure(x_range=country)

p1.vbar_stack(stackers=['TONS_HE', 'TONS_IC','TONS_FRAG'],
             x='COUNTRY_FLYING_MISSION', source=source,
             legend = ['High Explosive', 'Incendiary','Fragmentation'],
             width=0.5, color= Viridis[3]) # We mention the number of colors as 3 as we have 3 stacks. 
                                            # The number can be changed according to the number of stacks


p1.title.text ='Types of Munitions Dropped by Allied Country' #to give title to graph

# Labeling the graph
p1.xaxis.axis_label = 'Country'
p1.yaxis.axis_label = 'Kilotons of Munitions'

p1.xgrid.grid_line_color = None # Remove the x grid lines

# Adding interactive legend
p1.legend.location = "top_left"
p1.legend.click_policy="hide"

show(p1)
```

# Feature 2 : Hover Tool

```python

# Create the figure 
p2 = figure(x_range=country)

p2.vbar_stack(stackers=['TONS_HE', 'TONS_IC','TONS_FRAG'],
             x='COUNTRY_FLYING_MISSION', source=source,
             legend = ['High Explosive', 'Incendiary','Fragmentation'],
             width=0.5, color= Viridis[3]) # We mention the number of colors as 3 as we have 3 stacks. 
                                            # The number can be changed according to the number of stacks


p2.title.text ='Types of Munitions Dropped by Allied Country' #to give title to graph

# Labeling the graph
p2.xaxis.axis_label = 'Country'
p2.yaxis.axis_label = 'Kilotons of Munitions'

p2.xgrid.grid_line_color = None # Remove the x grid lines

# Adding interactive legend
p2.legend.location = "top_left"
p2.legend.click_policy="hide"

# Import Hover Tool
from bokeh.models import HoverTool

# Adding Hover Tool to the plot
hover = HoverTool()
hover.tooltips=[('Country: ', '@COUNTRY_FLYING_MISSION'),
    ('High Explosive: ', '@TONS_HE'),
    ('Incendiary: ', '@TONS_IC'),
    ('Fragmentation: ', '@TONS_FRAG')]

p2.add_tools(hover)

show(p2)
```

# Feature 3 : Creating multiple linked graphs

```python
data['Year'] = pd.DatetimeIndex(data['MSNDATE']).year
data['Month'] = pd.DatetimeIndex(data['MSNDATE']).month
data['Day'] = pd.DatetimeIndex(data['MSNDATE']).day
data.head()
```

```python
# Import gridplot
from bokeh.layouts import gridplot
output_notebook()

# Consider AC_ATTACKING as y, Year, Month and Day as Xs
linked_data_x = data.Year
linked_data_x1 = data.Month
linked_data_x2 = data.Day
linked_data_y = data.AC_ATTACKING

# Creating linked plot 1
linked_plot1 = figure(width=300, height=300)
linked_plot1.circle(linked_data_x, linked_data_y,fill_alpha=0.5)
#Adding labels to linked plot 1
linked_plot1.title.text = "Year wise Attacks"
linked_plot1.xaxis.axis_label = 'Year'
linked_plot1.yaxis.axis_label = 'Attacks'

# Creating linked plot 2
linked_plot2 = figure(width=300, height=300, y_range=linked_plot1.y_range)
linked_plot2.cross(linked_data_x1, linked_data_y)
linked_plot2.title.text = "Month wise attacks"
linked_plot2.xaxis.axis_label = 'Month'
linked_plot2.yaxis.axis_label = 'Attacks'

# Creating linked plot 3 
linked_plot3 = figure(width=300, height=300, y_range=linked_plot1.y_range)
linked_plot3.square(linked_data_x2, linked_data_y)
linked_plot3.title.text = "Day wise attacks"
linked_plot3.xaxis.axis_label = 'Day of the week'
linked_plot3.yaxis.axis_label = 'Attacks'

# Combining the subplots into a gridplot
p3 = gridplot([[linked_plot1, linked_plot2, linked_plot3]])


# show the results
show(p3)
```

# Feature 4 : Geographical Maps

```python
# Import
from pyproj import Proj, transform

AData=pd.read_csv("C:/Users/sushb/Downloads/DS861 Project/thor_wwii.csv")
AData.dropna(subset=['TAKEOFF_LONGITUDE','TAKEOFF_LATITUDE'],axis=0,inplace=True) #deleting the null values

def create_coordinates(long_arg,lat_arg):
    in_wgs = Proj(init='epsg:4326')
    out_mercator = Proj(init='epsg:3857')
    long, lat = long_arg, lat_arg
    mercator_x, mercator_y = transform(in_wgs, out_mercator, long, lat)
    return mercator_x,mercator_y

s=AData['TAKEOFF_LONGITUDE'].tolist()    
r=AData['TAKEOFF_LATITUDE'].tolist() 
x1 = []
y1 = []

x1,y1 = create_coordinates(s,r)

#print ("new value {} {}".format(x1,y1))

from bokeh.tile_providers import get_provider, Vendors

tile_provider = get_provider(Vendors.CARTODBPOSITRON)
#tile_provider = get_provider(Vendors.STAMEN_TONER_BACKGROUND)

# range bounds supplied in web mercator coordinates
p4 = figure(plot_width=800, 
           plot_height=400,
           x_range=(-12000000, 9000000), 
           y_range=(-1000000, 7000000), 
           x_axis_type='mercator', 
           y_axis_type='mercator')


p4.add_tile(tile_provider)

p4.circle(x=x1, y=y1, size=10, color='pink')

show(p4)
```
